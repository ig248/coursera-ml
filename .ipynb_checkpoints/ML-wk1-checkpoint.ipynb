### Linear regression with one variable

$m$ training examples: $(x^{(i)}, y^{(i)})$, $1\leq i\leq m$

Hypothesis: $$h_\theta(x) = \theta_0 + \theta_1 x$$

(shorthand $h(x)$)

`Training set` --> `Learning algorithm` --> `h`

`x` --> `h` --> `predicted y`

![](https://d3c33hcgiwev3.cloudfront.net/imageAssetProxy.v1/H6qTdZmYEeaagxL7xdFKxA_2f0f671110e8f7446bb2b5b2f75a8874_Screenshot-2016-10-23-20.14.58.png?expiry=1491868800000&hmac=S5H_baZUigbls-r5a6veaq0gQ5vzvFIooDy3_Camspc)

minimize *Cost function* (squared error function)
$$J(\theta_0, \theta_1)=\frac{1}{2m} \sum_{i=1}^m
\left(h_\theta(x^{(i)})-y^{(i)}\right)^2$$

---


### Gradient descent
![](https://d3c33hcgiwev3.cloudfront.net/imageAssetProxy.v1/bn9SyaDIEeav5QpTGIv-Pg_0d06dca3d225f3de8b5a4a7e92254153_Screenshot-2016-11-01-23.48.26.png?expiry=1491868800000&hmac=OJ0pPCZZR9BfHgCt8-jPo62gLbpxwdIiWkJAgfsD8lM)
Repeat until convergence:
$$\theta_j := \theta_j - \alpha \frac{\partial}{\partial \theta_j}J(\theta_0, \theta_1)$$

learning rate $\alpha$, *simultaneous update*

**Batch gradient descent**: using all training set for each step

````octave
# Linear Regression Excercise
# https://www.coursera.org/learn/machine-learning/exam/QeJ50/linear-regression-with-one-variable
x = [5 3 0 4];
y = [4 4 1 3];
t = [1, 0];
h = polyval(t, x)
m = length(x);
J = 1/(2*m)*sum((h-y).^2)
````

# Note on Octave on Ubuntu 16.04

`Ubuntu 16.04` install `Octave 4.0.0` by default.
The `submit()` function bugs out in this version.

Update to a later version from a PPA:
```
sudo apt-add-repository ppa:octave/stable
sudo apt-get update
sudo apt-get install octave
```

Building from source should also work, but I haven't tested it
